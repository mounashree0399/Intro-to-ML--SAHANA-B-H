```
from sklearn.preprocessing import StandardScaler
```

```
x1train = y_train.reshape(len(y_train),1)
x1test = y_test.reshape(len(y_test),1)
```

```
x_sc = StandardScaler()
y_sc = StandardScaler()
```

```
xx1_train = x_sc.fit_transform(x_train)
yx1_train = y_sc.fit_transform(x1train)
```

```
xx1_test = x_sc.transform(x_test)
yx1_test = y_sc.transform(x1test)
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures
```

```
r_reg = RandomForestRegressor()
l_reg = LinearRegression()
d_reg = DecisionTreeRegressor()
s_reg = SVR()
p_reg = LinearRegression()
```

```
x_pol = PolynomialFeatures(degree=1)
x_pol = x_pol.fit_transform(xx1_train)
```

```
temp = PolynomialFeatures(degree=1)
temp = temp.fit_transform(xx1_test)
```

```
r_reg.fit(xx1_train, yx1_train)
l_reg.fit(xx1_train, yx1_train)
d_reg.fit(xx1_train, yx1_train)
s_reg.fit(xx1_train, yx1_train)
p_reg.fit(x_pol, yx1_train)
```

```
r_pred = r_reg.predict(xx1_test)
l_pred = l_reg.predict(xx1_test)
d_pred = d_reg.predict(xx1_test)
s_pred = s_reg.predict(xx1_test)
p_pred = p_reg.predict(temp)
```

```
from sklearn.metrics import r2_score
```

```
r = r2_score(yx1_test, r_pred)
l = r2_score(yx1_test, l_pred)
d = r2_score(yx1_test, d_pred)
s = r2_score(yx1_test, s_pred)
p = r2_score(yx1_test, p_pred)
```

```
print(r, l, d, s, p)
```
